```python
import pandas as pd
import numpy as np
df= pd.read_csv("movie_dataset.csv")
df.head()
df.isna().sum()
```

```python
df['keywords'].replace(np.nan,'', inplace=True)
df['director'].replace(np.nan,'', inplace=True)
#,'genres','cast','director'
df['cast'].replace(np.nan,'', inplace=True)
df['genres'].replace(np.nan,'', inplace=True)
df.isna().sum()
```

```python
features=['keywords','cast','genres','director']
def combine_features(row):
        return row['keywords']+''+row['cast']+''+row['genres']+''+row['director']
#passing  a method to a df
df["combine_features"]=df.apply(combine_features, axis=1)
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
x=vec.fit_transform(df["combine_features"])
print(x.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_scr=cosine_similarity(x)
print (sim_scr)
```

```python
movie=input("Enter movie: ")
```

```python
sim_scr=-np.sort(-sim_scr)
#print(sim_scr)
def get_index(t):
    return df[df.title==t]["index"].values[0]
def get_title(i):
    return df[df.index==i]['title'].values[0]
movie_index=get_index(movie)
sim_movies=list(enumerate(sim_scr[movie_index]))
print(sim_movies)
```

```python
sorted_sim_movies=sorted(sim_movies,key=lambda x:x[1], reverse=True)
```

```python
count=0
for movie in sorted_sim_movies: 
    if count<15:
            print(get_title(movie[0]))
            count=count+1
```
